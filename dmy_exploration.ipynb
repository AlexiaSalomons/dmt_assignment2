```python
#Only keep the necessary features
training_data = training_data[['prop_id','visitor_location_country_id','prop_country_id','srch_booking_window','orig_destination_distance','price_usd',
                               'visitor_hist_adr_usd','visitor_hist_starrating','srch_destination_id','booking_bool','click_bool']]
```

```python
#Find the amount of NA and precentage of NA in dataset
total = training_data.isnull().sum().sort_values(ascending=False)
percent = (training_data.isnull().sum()/training_data.count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
#Set missing values to mean of the column
training_data['orig_destination_distance'].fillna(training_data['orig_destination_distance'].mean(), inplace=True)
training_data['visitor_hist_starrating'].fillna(training_data['visitor_hist_starrating'].mean(), inplace=True)
```

```python
#Check that there is no missing data
total = training_data.isnull().sum().sort_values(ascending=False)
percent = (training_data.isnull().sum()/training_data.count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
#Divide into 3 budget categories
training_data['visitor_hist_adr_usd'].quantile([0.3, 0.7])
```

```python
training_data['low_budget'] = (training_data['visitor_hist_adr_usd'] < 117.12)
training_data['medium_budget_temp_a'] = ((training_data['visitor_hist_adr_usd'] >= 117.12) & (training_data['visitor_hist_adr_usd'] < 200.32))
training_data['medium_budget_temp_b'] = (training_data['visitor_hist_adr_usd'].isna())
training_data['medium_budget'] = training_data['medium_budget_temp_a'] | training_data['medium_budget_temp_b']
training_data['high_budget'] = (training_data['visitor_hist_adr_usd'] >= 200.32)
```

```python
#Count the number of rows and number of bookings for every destination-hotel-budget combination.
group = training_data.groupby(['srch_destination_id','prop_id','low_budget','medium_budget','high_budget'])['booking_bool'].agg(['sum','count'])
```

```python
#Get the most popular hotel per destination and budget
weight_click = 0.05
agg = group.groupby(['srch_destination_id','prop_id','low_budget','medium_budget','high_budget']).sum().reset_index()
agg['count'] -= agg['sum']
agg = agg.rename(columns={'sum':'bookings','count':'clicks'})
agg['relevance'] = agg['bookings'] + weight_click * agg['clicks']
agg.head()
```

```python
def most_popular(group, n_max=10):
    relevance = group['relevance'].values
    prop_id = group['prop_id'].values
    most_popular = prop_id[np.argsort(relevance)[::-1]][:n_max]
    return np.array_str(most_popular)[1:-1] # remove square brackets
```

```python
most_pop = agg.groupby(['srch_destination_id','low_budget','medium_budget','high_budget']).apply(most_popular)
most_pop = pd.DataFrame(most_pop).rename(columns={0:'prop_id'})
most_pop.head()
```

```python
test_data = pd.read_csv('vu-dmt-assigment-2-2023/test_set_VU_DM.csv',dtype={'srch_destination_id':np.int32})
```

```python
test_data = test_data[['prop_id','visitor_location_country_id','prop_country_id','srch_booking_window','orig_destination_distance','price_usd',
                               'visitor_hist_adr_usd','visitor_hist_starrating','srch_destination_id']]
```

```python
test_data['low_budget'] = (test_data['visitor_hist_adr_usd'] < 117.12)
test_data['medium_budget_temp_a'] = ((test_data['visitor_hist_adr_usd'] >= 117.12) & (test_data['visitor_hist_adr_usd'] < 200.32))
test_data['medium_budget_temp_b'] = (test_data['visitor_hist_adr_usd'].isna())
test_data['medium_budget'] = test_data['medium_budget_temp_a'] | test_data['medium_budget_temp_b']
test_data['high_budget'] = (test_data['visitor_hist_adr_usd'] >= 200.32)
```

```python
test_data = test_data.merge(most_pop, how='left',left_on=['srch_destination_id','low_budget','medium_budget','high_budget'],right_index=True)
test_data.head()

test_data.prop_id.to_csv('submission.csv',header=True, index_label='id')
```
